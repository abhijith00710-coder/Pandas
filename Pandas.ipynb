```python
import pandas as pd
import numpy as np
a=np.arange(4,10)

s=pd.Series(a)

print(s)

print(s.ndim)
print(s.shape)
print(s.size)

#for getting initial rows up to range

print(s.head) # normally 5
print(s.head(4))

# for getting funal rows

print(s.tail(3))
```

```python
# create two series

a=pd.Series([4,5,6,7,8])

b=pd.Series([5,4,3,6,7])

# print(a)
# print(b)

# conactenate two series

d=pd.concat([a,b],ignore_index=True) # ignore index koduthillel pazhaya index
                                     # show chyum athil
print(d)
```

```python
# Adding two series

e=a.add(b)
print(e)
```

```python
# Subtracting two series

e=a.sub(b)
print(e)
```

```python
# Multiplyng two series

e=a.mul(b)
print(e)
```

```python
# dividing two series

e=a.div(b)
print(e)
```

```python
# for custom indexing use , index function and mention the custom index you want to print

z=pd.Series([3,4,5,6],index=["a","b","c","d"])
print(z)
```

```python
d={"id":1,"age":54,"class":"ds"}

z=pd.Series(d)
print(z)  # here key become index of that
```

```python
z=pd.Series(4,index=[1,2,3,4])
z
```

```python
z=pd.Series([1,2,3,4,5,6,7])

print(z)

z[3]

z[1:4]

z[[6,1,5]]  # seperate index mathram edukkan coma ettu double
            # closed bracket il aakkuka
```

```python
import pandas as pd

# creating data frame from list of dictionaries

l=[{"fruits":"apple","price":234},{"fruits":"orange","price":87},
 {"fruits":"grapes","price":90}]

df=pd.DataFrame(l)

df
```
